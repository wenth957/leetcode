# 1 数据分析流程
- 数据是对现实世界现象的观测，存在错误、冗余、缺失
- 任务是根据数据进行决策，充满了错误的开始和死胡同。大致流程如下
  - 汤生路透等金融机构收集数据汇总到数据库中
  - 被某个公司买去，转换为一个Hadoop集群上的Hive仓库
  - 利用某个脚本读取数据，进行清洗，导出为另一个文件
  - 使用python等程序进行建模分析，预测结果储存为一个文件
  - 使用估值建模方法对股价进行预测
  - 模型多次迭代后，由产品团队用C++重写
  - 全数据上运行，结果输出到另一个数据库中
- 模型：数据可以直接输入模型吗？一般需要转变成数值型，这就是特征。并不是所有特征都适合某个模型
- 特征工程：设计、选出合适的特征
- 模型评价：特征对模型复杂度的影响，好的特征使模型更简单


# 2 数值
- 特征的量纲：标准化，树模型使用的是阶梯函数所以不受量纲的影响
- 特征的分布：线性回归，回归有多个结果，使用对数消除异方差
- 特征的组合：更加简洁的捕获原始数据中的信息

## 2.1 特征空间
- 标量：单一的数值型
- 向量：数据的有序列表，有实际意义
- 向量空间：特征空间，所有数据的集合，点云

## 2.2 计数数据
- 访问量、收听等，可能会有极端值，保留原始数据还是转化为其他数据
- 模型可能会被异常值带偏
- 目标变量的选择：对歌曲的喜爱不能只用原始数据来度量，应该转化为其他数据类型

## 2.3 分箱（区间量化）

- 会改变分布
- 点评数量和商家高评分
- 点评数量使用原始数据还是进一步处理
- 数据中具有异常值，会干扰相似度的度量
- 均匀化分布：分箱重新定义变量数值，使得分布更加均匀
- 人工分组、固定宽度、指数宽度(对数变换、量级相差较大)、分位数分箱、Box-Cox变换

## 2.4 特征缩放/归一化
- 不会改变分布
- min-max
- 标准化
- 以上两种方法不适合稀疏数据：0占大多数，以上处理会使数据向量密集化
- l2归一化，x/||x|| , 缩放到（-1，1）之间  ，也可以对特征空间进行归一化
- 解决尺度差别

## 2.5 交互特征
- 同时满足两个特征的样本
- 交互项的引入，PolynomialFeatures
- 计算成本，特征选择

## 2.6 特征选择
- 简约模型，提高计算速度
- 过滤技术：相关性，阈值，谨慎使用
- 打包方法：实验各个特征子集，根据得分选取特征
- 嵌入式方法：决策树、正则化
