```python
# 1.5贴吧爬虫
import requests
class TiebaSpider:
     def __init__(self,tieba_name):
         self.tieba_name = tieba_name
         # 在字符串中传入参数
         self.url_temp = 'https://tieba.baidu.com/f?kw='+tieba_name+'&ie=utf-8&pn={}'
         self.headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.72 Safari/537.36 Edg/90.0.818.42'}
     
     def get_url_list(self):
         # 构造url列表
         url_list = []
         for i in range(1000):
             url_list.append(self.url_temp.format(i*50))
         return url_list
     
     def parse_url(self,url):
         #发送请求，获得html字符串
         r = requests.get(url,headers=self.headers)
         return r.content.decode()
    
     def save_html(self,html_str,page_num):
         # 保存本地
         file_path = '{}-第{}页.html'.format(self.tieba_name,page_num)
         with open(file_path,'w',encoding='utf-8') as f:# 李毅第i页.html
             f.write(html_str)

     def run(self):
         # 实现主要逻辑：只调用相应方法
         # 1 构造url列表
         url_list = self.get_url_list()
         # 2 遍历，发送请求，获取响应
         for url in url_list:
             html_str=self.parse_url(url)
             # 3 保存页面到本地
             page_num = url_list.index(url)+1 
             self.save_html(html_str,page_num)
if   __name__=='__main__':
        tieba_spiders = TiebaSpider('李毅')
        tieba_spiders.run()
```

```python
# 配合状态码判断是否请求成功
import requests

def _parse_url(url):
    r = requests.get(url,headers=headers,timeout=3)
    assert r.status_code == 200
    return r.content.decode()

def parse_url(url):
    try:
        html_str = _parse_url(url)
    except:
        html_str = None
    
    return html_str

```
